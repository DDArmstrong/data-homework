## 2) What's the current wind speed, and how much warmer does it feel than it actually is?

- *Tip: You can do this by browsing through the dictionaries, but it might be easier to read the documentation*
- *Tip: For the second half: it **is** one temperature, and it **feels** a different temperature. Calculate the difference. Same as we did last time!*

From the documentation at https://www.weatherapi.com/docs/:

Field		Data Type	Description	
temp_c		decimal		Temperature in celsius	
temp_f		decimal		Temperature in fahrenheit	
feelslike_c	decimal		Feels like temperature in celsius	
feelslike_f	decimal		Feels like temperature in fahrenheit	
wind_mph	decimal		Wind speed in miles per hour	
wind_kph	decimal		Wind speed in kilometer per hour	

```python
response = urllib.request.urlopen(url)
obj = json.load(response)
obj
```

```python
wind = (obj['current']['wind_mph'])
actual_temp = (obj['current']['temp_f'])
perceived_temp = (obj['current']['feelslike_f'])
difference = perceived_temp - actual_temp
print("In Yazoo City MS, the wind is blowing at","{:.{}f}".format( wind, 1 ),"mph.")
print("It's", "{:.{}f}".format( actual_temp, 1 ),"degrees.")
print("It feels like it's", "{:.{}f}".format( perceived_temp, 1 ),"degrees.")
print("That's a difference of","{:.{}f}".format( difference, 1 ),"degrees.")

"{:.{}f}".format( actual_temp, 1 )

# print(obj.get('current'))
# dict.keys(obj['current'])
```
